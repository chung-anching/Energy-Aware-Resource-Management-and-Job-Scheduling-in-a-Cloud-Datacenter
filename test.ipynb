```python
import pandas as pd
import numpy as np
from datetime import timedelta
from typing import Dict, List, Tuple
import random
import matplotlib.pyplot as plt

# Constants from the paper
JOB_TYPES = ["AR", "IM", "BE"]
JOB_DISTRIBUTION = {"AR": 0.2, "IM": 0.3, "BE": 0.5}
IDLE_THRESHOLD_MINUTES = 30
MIN_IDLE_HOSTS = 1

# Function to classify jobs based on resource usage with random distribution
def classify_jobs(cpu_load: float, job_dist: Dict) -> List[Tuple[str, float]]:
    if cpu_load <= 0:
        return []
    job_types = list(job_dist.keys())
    probabilities = list(job_dist.values())
    n_trials = 1000
    random_loads = np.random.multinomial(n_trials, probabilities) / n_trials * cpu_load
    jobs = [(job_type, max(0.001, load)) for job_type, load in zip(job_types, random_loads)]
    print("jobsjobs", jobs)
    return jobs  # [('AR', 0.65464), ('IM', 0.9752799999999999), ('BE', 1.71008)]

# Battery Manager for AgentBAT
class BatteryManager:
    def __init__(self, capacity: float = 1000.0):  # Capacity in kWh
        self.capacity = capacity
        self.current_level = 0.0  # Current battery level in kWh
        self.max_charge_rate = 100.0  # kW (max charge/discharge rate per hour)
        self.charge_efficiency = 0.95  # Charge efficiency
        self.discharge_efficiency = 0.95  # Discharge efficiency

    def charge(self, power: float, carbon_intensity: float) -> float:
        if carbon_intensity < 0.55:  # Charge when CI is low
            charge_amount = min(self.max_charge_rate, power, (self.capacity - self.current_level) / self.charge_efficiency)
            self.current_level += charge_amount * self.charge_efficiency
            return charge_amount
        return 0.0

    def discharge(self, power_needed: float, carbon_intensity: float) -> float:
        if carbon_intensity >= 0.55:  # Discharge when CI is high
            discharge_amount = min(self.max_charge_rate, power_needed, self.current_level / self.discharge_efficiency)
            self.current_level -= discharge_amount / self.discharge_efficiency
            return discharge_amount
        return 0.0

    def get_level(self) -> float:
        return self.current_level

# Node class with waiting time tracking
class Node:
    def __init__(self, name: str):
        self.name = name
        self.state = "Idle"
        self.cpu_used = 0
        self.gpu_power_used = 0
        self.gpu_mem_used = 0
        self.vm_queue = []
        self.last_active_time = None
        self.available_time = 0

    def update_state(self, time_minutes: float, job_load: float = 0):
        if job_load > 0 and self.state in ["Idle", "Standby"]:
            self.state = "Active"
            self.last_active_time = time_minutes
        elif job_load == 0 and self.state == "Active":
            if time_minutes - self.last_active_time > IDLE_THRESHOLD_MINUTES:
                self.state = "Standby"
            else:
                self.state = "Idle"
        self.cpu_used = job_load
        if NODE_SPEC[self.name]["gpu_count"] > 0:
            self.gpu_power_used = job_load * 10

    def get_power_consumption(self) -> float:
        if self.state == "Active":
            return NODE_SPEC[self.name]["power_active"]
        elif self.state == "Idle":
            return NODE_SPEC[self.name]["power_idle"]
        else:
            return NODE_SPEC[self.name]["power_standby"]

    def can_handle_job(self, cpu_load: float, gpu_power: float, gpu_mem: float) -> bool:
        cpu_capacity = NODE_SPEC[self.name]["cpu_cores"] * 100
        gpu_capacity = NODE_SPEC[self.name]["gpu_count"] * 100
        return (self.cpu_used + cpu_load <= cpu_capacity and 
                (NODE_SPEC[self.name]["gpu_count"] == 0 or 
                 (self.gpu_power_used + gpu_power <= gpu_capacity and 
                  self.gpu_mem_used + gpu_mem <= NODE_SPEC[self.name]["gpu_memory_gb"] * 1024)))

    def reset_usage(self):
        self.cpu_used = 0
        self.gpu_power_used = 0
        self.gpu_mem_used = 0
        self.vm_queue = []
        self.state = "Idle"
        self.last_active_time = None
        self.available_time = 0

# Scheduling Algorithms with Solar and Carbon Integration
def eavmat_scheduling(nodes: List[Node], workload: pd.DataFrame, solar_watt_list: List[float], hourly_carbon_list: List[float], time_index: List[str], battery: BatteryManager) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {},
        "battery_level_over_time": {},
        "waiting_times": []  # Store waiting times for each job
    }
    total_jobs = 0
    successful_jobs = 0

    times = workload["hour"].unique()
    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum()
        results["energy_consumption_over_time"][time] = 0
        results["battery_level_over_time"][time] = battery.get_level()

    for idx, row in workload.iterrows():
        time = pd.to_datetime(row["hour"])
        time_str = str(time)
        time_idx = list(time_index).index(time_str) if time_str in time_index else 0
        solar_power = solar_watt_list[time_idx] / 1000  # Convert Watt to kW
        carbon_intensity = hourly_carbon_list[time_idx]

        # Adjust active nodes based on solar power
        active_node_limit = max(MIN_IDLE_HOSTS, int(len(nodes) * (solar_power / max(solar_watt_list))))
        active_nodes = [n for n in nodes if n.state in ["Active", "Idle"]][:active_node_limit]

        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        total_jobs += len(jobs)

        node = next((n for n in nodes if n.name == nodename), None)
        if node is None:
            continue

        for job_type, load in jobs:
            scheduled = False
            start_time = time
            scheduled_time = None

            if node in active_nodes and node.can_handle_job(load, gpu_power, gpu_mem):
                node.cpu_used += load
                if NODE_SPEC[nodename]["gpu_count"] > 0:
                    node.gpu_power_used += gpu_power
                    node.gpu_mem_used += gpu_mem
                node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                scheduled = True
                scheduled_time = time
                successful_jobs += 1
            else:
                other_nodes = [n for n in active_nodes if n.name != nodename and n.state in ["Active", "Idle"]]
                for other_node in sorted(other_nodes, key=lambda x: x.available_time):
                    if other_node.can_handle_job(load, gpu_power, gpu_mem):
                        other_node.cpu_used += load
                        if NODE_SPEC[other_node.name]["gpu_count"] > 0:
                            other_node.gpu_power_used += gpu_power
                            other_node.gpu_mem_used += gpu_mem
                        other_node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                        other_node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                        other_node.state = "Active"
                        scheduled = True
                        scheduled_time = time
                        successful_jobs += 1
                        break

            if scheduled:
                waiting_time = (scheduled_time - start_time).total_seconds() / 60 if scheduled_time else 0
                results["waiting_times"].append(waiting_time)
                results["node_states_over_time"][time][node.name] = node.state
                power_consumption = node.get_power_consumption() / 1000  # Convert Watt to kW
                battery_discharge = battery.discharge(power_consumption, carbon_intensity)
                power_needed = power_consumption - battery_discharge
                battery_charge = battery.charge(solar_power, carbon_intensity)
                results["energy_consumption_over_time"][time] += power_needed * 60  # kW to kWh
                results["battery_level_over_time"][time] = battery.get_level()

    total_time_minutes = 60
    for time in times:
        for node in nodes:
            state_time = total_time_minutes if results["node_states_over_time"][time][node.name] == "Active" else 0
            results["energy_consumption"] += node.get_power_consumption() * state_time / len(times)

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])
    results["average_waiting_time"] = np.mean(results["waiting_times"]) if results["waiting_times"] else 0

    return results

# Similar modifications for other scheduling algorithms (PSO, Round Robin, SJF, FCFS)
def pso_scheduling(nodes: List[Node], workload: pd.DataFrame, solar_watt_list: List[float], hourly_carbon_list: List[float], time_index: List[str], battery: BatteryManager, iterations: int = 10, particles: int = 5) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {},
        "battery_level_over_time": {},
        "waiting_times": []
    }
    total_jobs = 0
    successful_jobs = 0

    def fitness(particle):
        energy = 0
        for job_node in particle:
            node = nodes[job_node[1]]
            if node.can_handle_job(job_node[0][1], job_node[0][2], job_node[0][3]):
                energy += node.get_power_consumption()
        return -energy

    jobs = []
    times = workload["hour"].unique()
    for idx, row in workload.iterrows():
        time = pd.to_datetime(row["hour"])
        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        classified_jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        for job_type, load in classified_jobs:
            jobs.append((job_type, load, gpu_power, gpu_mem, time, nodename))

    total_jobs = len(jobs)

    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum()
        results["energy_consumption_over_time"][time] = 0
        results["battery_level_over_time"][time] = battery.get_level()

    particle_positions = [[(job, random.randint(0, len(nodes)-1)) for job in jobs] for _ in range(particles)]
    particle_velocities = [[0 for _ in range(len(jobs))] for _ in range(particles)]
    personal_best = particle_positions.copy()
    personal_best_fitness = [fitness(p) for p in particle_positions]
    global_best = particle_positions[np.argmax(personal_best_fitness)]
    global_best_fitness = max(personal_best_fitness)

    for _ in range(iterations):
        for i in range(particles):
            for j in range(len(jobs)):
                inertia = 0.7
                cognitive = 0.5
                social = 0.5
                r1, r2 = random.random(), random.random()
                particle_velocities[i][j] = (inertia * particle_velocities[i][j] + 
                                            cognitive * r1 * (personal_best[i][j][1] - particle_positions[i][j][1]) + 
                                            social * r2 * (global_best[j][1] - particle_positions[i][j][1]))
                new_position = particle_positions[i][j][1] + particle_velocities[i][j]
                print("new_position", new_position)
                # new_position = max(0, min(len(nodes)-1), int(new_position))
                new_position = max(0, min(len(nodes) - 1, int(new_position)))
                particle_positions[i][j] = (particle_positions[i][j][0], new_position)

            current_fitness = fitness(particle_positions[i])
            if current_fitness > personal_best_fitness[i]:
                personal_best[i] = particle_positions[i].copy()
                personal_best_fitness[i] = current_fitness
                if current_fitness > global_best_fitness:
                    global_best = particle_positions[i].copy()
                    global_best_fitness = current_fitness

    for job, node_idx in global_best:
        time = job[4]
        time_str = str(time)
        time_idx = list(time_index).index(time_str) if time_str in time_index else 0
        solar_power = solar_watt_list[time_idx] / 1000
        carbon_intensity = hourly_carbon_list[time_idx]

        active_node_limit = max(MIN_IDLE_HOSTS, int(len(nodes) * (solar_power / max(solar_watt_list))))
        if node_idx >= active_node_limit:
            continue

        node = nodes[node_idx]
        if node.can_handle_job(job[1], job[2], job[3]):
            node.cpu_used += job[1]
            if NODE_SPEC[node.name]["gpu_count"] > 0:
                node.gpu_power_used += job[2]
                node.gpu_mem_used += job[3]
            node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, job[1])
            node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
            waiting_time = 0  # PSO assumes immediate scheduling after optimization
            results["waiting_times"].append(waiting_time)
            successful_jobs += 1
            results["node_states_over_time"][time][node.name] = node.state
            power_consumption = node.get_power_consumption() / 1000
            battery_discharge = battery.discharge(power_consumption, carbon_intensity)
            power_needed = power_consumption - battery_discharge
            battery_charge = battery.charge(solar_power, carbon_intensity)
            results["energy_consumption_over_time"][time] += power_needed * 60
            results["battery_level_over_time"][time] = battery.get_level()

    total_time_minutes = 60
    for time in times:
        for node in nodes:
            state_time = total_time_minutes if results["node_states_over_time"][time][node.name] == "Active" else 0
            results["energy_consumption"] += node.get_power_consumption() * state_time / len(times)

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])
    results["average_waiting_time"] = np.mean(results["waiting_times"]) if results["waiting_times"] else 0

    return results

# Add similar modifications to round_robin_scheduling, sjf_scheduling, and fcfs_scheduling

# Visualization with battery level
def visualize_results(all_results):
    times = list(all_results["EAVMAT"]["node_states_over_time"].keys())
    times = pd.to_datetime(times)

    # 1. Node State Changes Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        state_nums = []
        for time in times:
            states = list(result["node_states_over_time"][time].values())
            active_count = sum(1 for s in states if s == "Active")
            idle_count = sum(1 for s in states if s == "Idle")
            standby_count = sum(1 for s in states if s == "Standby")
            total_nodes = len(states)
            state_avg = (active_count * 1 + standby_count * 2) / total_nodes if total_nodes > 0 else 0
            state_nums.append(state_avg)
        plt.plot(times, state_nums, label=f"{algo_name} Node State", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Node State (0=Idle, 1=Active, 2=Standby)')
    plt.title('Node State Changes Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 2. CPU Load Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        cpu_loads = [result["cpu_load_over_time"][t] for t in times]
        plt.plot(times, cpu_loads, label=f"{algo_name} CPU Load", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Total CPU Load')
    plt.title('CPU Load Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 3. Energy Consumption Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        energy_values = [result["energy_consumption_over_time"][t] for t in times]
        plt.plot(times, energy_values, label=f"{algo_name} Energy", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Energy Consumption (Watt-minutes)')
    plt.title('Energy Consumption Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 4. Battery Level Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        battery_levels = [result["battery_level_over_time"][t] for t in times]
        plt.plot(times, battery_levels, label=f"{algo_name} Battery Level", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Battery Level (kWh)')
    plt.title('Battery Level Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    # Initialize nodes (NODE_SPEC should be defined as in previous examples)
    nodes = [Node(name) for name in NODE_SPEC.keys()]


    workload = CPU_request_data[0:100]
    workload["hour"] = pd.to_datetime(workload["hour"])

    # Load new data
    # solar_watt_list = np.array([0.018, 0.018, 0.018, 0.018, 0.018, 0.1674, 1.316, 3.357, 1.317, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 1.055, 1.958, 0.981, 0.069, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.0577, 1.479, 0.503, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.897, 0.472, 0.805, 0.348, 0.055, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.0132, 0.567, 1.025, 2.25, 0.89, 0.992, 0.552, 0.966, 0.018, 0.374, 0.364, 0.1, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.0159, 0.0429, 0.118, 0.176, 0.355, 0.341, 0.412, 0.7677, 0.628, 0.539, 0.39, 0.374, 0.164, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.0135, 0.254, 0.7427, 1.759, 1.398, 0.9071, 0.178, 1.3337, 0.6957, 0.7314, 0.9917, 0.1453])
    # hourly_carbon_list = np.array([0.56992881, 0.57394017, 0.57765505, 0.54256068, 0.53978316, 0.53808131])
    # CPU_request_time = np.array(['2024-11-05T00:00:00.000000000', '2024-11-05T01:00:00.000000000', '2024-11-05T02:00:00.000000000', '2024-12-31T06:00:00.000000000', '2024-12-31T07:00:00.000000000', '2024-12-31T08:00:00.000000000'])

    # Initialize battery
    battery = BatteryManager()

    # Run and compare all scheduling algorithms
    algorithms = {
        "EAVMAT": lambda n, w: eavmat_scheduling(n, w, solar_watt_list, hourly_carbon_list, CPU_request_time, battery),
        "PSO": lambda n, w: pso_scheduling(n, w, solar_watt_list, hourly_carbon_list, CPU_request_time, battery),
        # Add other algorithms similarly
    }

    all_results = {}
    for name, algo in algorithms.items():
        for node in nodes:
            node.reset_usage()
        battery.current_level = 0.0  # Reset battery for each algorithm
        results = algo(nodes.copy(), workload)
        all_results[name] = results
        print(f"\n{name} Scheduling Results:")
        print(f"Total Energy Consumption (Watt-minutes): {results['energy_consumption']}")
        print(f"Job Success Rate (%): {results['job_success_rate']:.2f}")
        print(f"Number of Active Hosts: {results['active_hosts']}")
        print(f"Average Waiting Time (minutes): {results['average_waiting_time']:.2f}")
        for node in nodes:
            if node.state == "Active":
                print(f"Node {node.name} - State: {node.state}, CPU Used: {node.cpu_used}, GPU Power Used: {node.gpu_power_used}, Power: {node.get_power_consumption()} Watts")

    # Visualize results
    visualize_results(all_results)
```
