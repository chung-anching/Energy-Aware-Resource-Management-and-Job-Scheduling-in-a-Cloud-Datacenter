```python
import pandas as pd
import numpy as np
from datetime import timedelta
from typing import Dict, List, Tuple
import random
import matplotlib.pyplot as plt

# Constants from the paper
JOB_TYPES = ["AR", "IM", "BE"]  # Advance Reservation, Immediate, Best Effort
JOB_DISTRIBUTION = {"AR": 0.2, "IM": 0.3, "BE": 0.5}  # Percentage of each job type
IDLE_THRESHOLD_MINUTES = 30  # I_th
MIN_IDLE_HOSTS = 1

# Function to classify jobs based on resource usage with random distribution
def classify_jobs(cpu_load: float, job_dist: Dict) -> List[Tuple[str, float]]:
    if cpu_load <= 0:
        return []
    job_types = list(job_dist.keys())
    probabilities = list(job_dist.values())
    n_trials = 1000
    random_loads = np.random.multinomial(n_trials, probabilities) / n_trials * cpu_load
    jobs = [(job_type, max(0.001, load)) for job_type, load in zip(job_types, random_loads)]
    print("jobsjobs", jobs)
    return jobs  # [('AR', 0.65464), ('IM', 0.9752799999999999), ('BE', 1.71008)]

# Node class to manage state and resources
class Node:
    def __init__(self, name: str):
        self.name = name
        self.state = "Idle"
        self.cpu_used = 0
        self.gpu_power_used = 0
        self.gpu_mem_used = 0
        self.vm_queue = []
        self.last_active_time = None
        self.available_time = 0

    def update_state(self, time_minutes: float, job_load: float = 0):
        if job_load > 0 and self.state in ["Idle", "Standby"]:
            self.state = "Active"
            self.last_active_time = time_minutes
        elif job_load == 0 and self.state == "Active":
            if time_minutes - self.last_active_time > IDLE_THRESHOLD_MINUTES:
                self.state = "Standby"
            else:
                self.state = "Idle"
        self.cpu_used = job_load
        if NODE_SPEC[self.name]["gpu_count"] > 0:
            self.gpu_power_used = job_load * 10

    def get_power_consumption(self) -> float:
        if self.state == "Active":
            return NODE_SPEC[self.name]["power_active"]
        elif self.state == "Idle":
            return NODE_SPEC[self.name]["power_idle"]
        else:
            return NODE_SPEC[self.name]["power_standby"]

    def can_handle_job(self, cpu_load: float, gpu_power: float, gpu_mem: float) -> bool:
        cpu_capacity = NODE_SPEC[self.name]["cpu_cores"] * 2
        gpu_capacity = NODE_SPEC[self.name]["gpu_count"] * 2
        return (self.cpu_used + cpu_load <= cpu_capacity and 
                (NODE_SPEC[self.name]["gpu_count"] == 0 or 
                 (self.gpu_power_used + gpu_power <= gpu_capacity and 
                  self.gpu_mem_used + gpu_mem <= NODE_SPEC[self.name]["gpu_memory_gb"] * 1024)))

    def reset_usage(self):
        self.cpu_used = 0
        self.gpu_power_used = 0
        self.gpu_mem_used = 0
        self.vm_queue = []
        self.state = "Idle"
        self.last_active_time = None
        self.available_time = 0

# Scheduling Algorithms
def eavmat_scheduling(nodes: List[Node], workload: pd.DataFrame) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {}
    }
    total_jobs = 0
    successful_jobs = 0

    times = workload["hour"].unique()
    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum()  # 全部node賈總
        results["energy_consumption_over_time"][time] = 0

    for idx, row in workload.iterrows(): # 輪流每個node
        time = pd.to_datetime(row["hour"])
        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        total_jobs += len(jobs)

        node = next((n for n in nodes if n.name == nodename), None) # 它会遍历 nodes 中的每一个元素 n，然后检查每个元素的 name 属性是否等于 nodename。如果条件成立，就会生成这个元素。
        if node is None:
            continue

        for job_type, load in jobs: # [('AR', 0.65464), ('IM', 0.9752799999999999), ('BE', 1.71008)]
            scheduled = False
            if node.can_handle_job(load, gpu_power, gpu_mem):
                node.cpu_used += load
                if NODE_SPEC[nodename]["gpu_count"] > 0:
                    node.gpu_power_used += gpu_power
                    node.gpu_mem_used += gpu_mem
                node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                scheduled = True
                successful_jobs += 1
            else:
                # If this node can't handle, try other active/idle nodes (simplified)
                other_nodes = [n for n in nodes if n.name != nodename and n.state in ["Active", "Idle"]]
                for other_node in sorted(other_nodes, key=lambda x: x.available_time):
                    if other_node.can_handle_job(load, gpu_power, gpu_mem):
                        other_node.cpu_used += load
                        if NODE_SPEC[other_node.name]["gpu_count"] > 0:
                            other_node.gpu_power_used += gpu_power
                            other_node.gpu_mem_used += gpu_mem
                        other_node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                        other_node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                        other_node.state = "Active"
                        scheduled = True
                        successful_jobs += 1
                        break

            if not scheduled and job_type in ["AR", "IM"]:
                for target_node in [node] + other_nodes:
                    if target_node.vm_queue and any(j[0] == "BE" for j in target_node.vm_queue): # 如果队列中有后台作业，可以尝试进行抢占。
                        target_node.vm_queue = [j for j in target_node.vm_queue if j[0] != "BE"] # 将 target_node 的 vm_queue 更新为一个新的队列，删除所有类型为 "BE" 的作业。即通过删除后台作业，将资源重新分配给前台作业
                        target_node.cpu_used += load
                        if NODE_SPEC[target_node.name]["gpu_count"] > 0:
                            target_node.gpu_power_used += gpu_power
                            target_node.gpu_mem_used += gpu_mem
                        target_node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load) # 调用 update_state() 方法更新目标节点的状态。这个方法的参数是一个时间差，表示从某个特定时间点（2024年11月4日15:00）到当前时间 time 的分钟数。load 表示新增的负载。
                        target_node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60 # 将目标节点的 available_time 设置为当前时间 time 加 1 小时后的时间差（以分钟为单位）。这意味着这个节点在未来 1 小时内将不再可用，直到时间过去。
                        scheduled = True
                        successful_jobs += 1
                        break

            if scheduled:
                results["node_states_over_time"][time][node.name] = node.state
                results["energy_consumption_over_time"][time] += node.get_power_consumption()

    total_time_minutes = 60
    for time in times: # 外层循环遍历所有的时间点 times。
        for node in nodes: # 内层循环遍历所有的节点 nodes。
            state_time = total_time_minutes if results["node_states_over_time"][time][node.name] == "Active" else 0
            results["energy_consumption"] += node.get_power_consumption() * state_time / len(times)

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])
    print("resultsresults", results) # {'energy_consumption': 0.0, 'job_success_rate': 0.0, 'active_hosts': 0, 'node_states_over_time': {Timestamp('2024-11-04 14:00:00'): {'node-171': 'Idle', 'node-172': 'Idle', 'esxi-37': 'Idle'}, Timestamp('2024-11-04 16:00:00'): {'node-161': 'Idle', 'node-162': 'Idle', }, Timestamp('2024-11-04 17:00:00'): {'node-32': 'Idle', 'esxi-37': 'Idle'}, Timestamp('2024-11-04 18:00:00'): {'node-32': 'Idle', 'esxi-37': 'Idle'}}, 'cpu_load_over_time': {Timestamp('2024-11-04 14:00:00'): 338.745, Timestamp('2024-11-04 15:00:00'): 347.4925, Timestamp('2024-11-04 16:00:00'): 386.78, Timestamp('2024-11-04 17:00:00'): 415.544166666667, Timestamp('2024-11-04 18:00:00'): 316.81916666666666}, 'energy_consumption_over_time': {Timestamp('2024-11-04 14:00:00'): 0, Timestamp('2024-11-04 15:00:00'): 0, Timestamp('2024-11-04 16:00:00'): 0, Timestamp('2024-11-04 17:00:00'): 0, Timestamp('2024-11-04 18:00:00'): 0}}

    return results

def pso_scheduling(nodes: List[Node], workload: pd.DataFrame, iterations: int = 10, particles: int = 5) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {}
    }
    total_jobs = 0
    successful_jobs = 0

    def fitness(particle):
        energy = 0
        for job_node in particle:
            node = nodes[job_node[1]]
            if node.can_handle_job(job_node[0][1], job_node[0][2], job_node[0][3]):
                energy += node.get_power_consumption()
        return -energy

    jobs = []
    times = workload["hour"].unique()
    for idx, row in workload.iterrows():
        time = pd.to_datetime(row["hour"])
        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        classified_jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        for job_type, load in classified_jobs:
            jobs.append((job_type, load, gpu_power, gpu_mem, time, nodename))

    total_jobs = len(jobs)

    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum() # 每個演算法數字都一樣
        results["energy_consumption_over_time"][time] = 0

    particle_positions = [[(job, random.randint(0, len(nodes)-1)) for job in jobs] for _ in range(particles)]
    particle_velocities = [[0 for _ in range(len(jobs))] for _ in range(particles)]
    personal_best = particle_positions.copy()
    personal_best_fitness = [fitness(p) for p in particle_positions]
    global_best = particle_positions[np.argmax(personal_best_fitness)]
    global_best_fitness = max(personal_best_fitness)

    for _ in range(iterations):
        for i in range(particles):
            for j in range(len(jobs)):
                inertia = 0.7
                cognitive = 0.5
                social = 0.5
                r1, r2 = random.random(), random.random()
                particle_velocities[i][j] = (inertia * particle_velocities[i][j] + 
                                            cognitive * r1 * (personal_best[i][j][1] - particle_positions[i][j][1]) + 
                                            social * r2 * (global_best[j][1] - particle_positions[i][j][1]))
                new_position = particle_positions[i][j][1] + particle_velocities[i][j]
                new_position = max(0, min(len(nodes)-1, int(new_position)))
                particle_positions[i][j] = (particle_positions[i][j][0], new_position)

            current_fitness = fitness(particle_positions[i])
            if current_fitness > personal_best_fitness[i]:
                personal_best[i] = particle_positions[i].copy()
                personal_best_fitness[i] = current_fitness
                if current_fitness > global_best_fitness:
                    global_best = particle_positions[i].copy()
                    global_best_fitness = current_fitness

    for job, node_idx in global_best:
        node = nodes[node_idx]
        if node.can_handle_job(job[1], job[2], job[3]):
            node.cpu_used += job[1]
            if NODE_SPEC[node.name]["gpu_count"] > 0:
                node.gpu_power_used += job[2]
                node.gpu_mem_used += job[3]
            time = job[4]
            node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, job[1])
            node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
            successful_jobs += 1
            results["node_states_over_time"][time][node.name] = node.state
            results["energy_consumption_over_time"][time] += node.get_power_consumption()

    total_time_minutes = 60
    for time in times:
        for node in nodes:
            state_time = total_time_minutes if results["node_states_over_time"][time][node.name] == "Active" else 0
            results["energy_consumption"] += node.get_power_consumption() * state_time / len(times)
            results["energy_consumption_over_time"][time] += node.get_power_consumption() * (state_time / len(times)) if results["node_states_over_time"][time][node.name] != "Idle" else 0

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])

    return results

def round_robin_scheduling(nodes: List[Node], workload: pd.DataFrame) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {}
    }
    total_jobs = 0
    successful_jobs = 0

    jobs = []
    times = workload["hour"].unique()
    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum()
        results["energy_consumption_over_time"][time] = 0

    for idx, row in workload.iterrows():
        time = pd.to_datetime(row["hour"])
        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        classified_jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        for job_type, load in classified_jobs:
            jobs.append((job_type, load, gpu_power, gpu_mem, time, nodename))

    total_jobs = len(jobs)
    node_idx = 0

    for job in jobs:
        job_type, load, gpu_power, gpu_mem, time, original_node = job
        for _ in range(len(nodes)):
            if node_idx >= len(nodes):
                node_idx = 0
            node = nodes[node_idx]
            if node.can_handle_job(load, gpu_power, gpu_mem):
                node.cpu_used += load
                if NODE_SPEC[node.name]["gpu_count"] > 0:
                    node.gpu_power_used += gpu_power
                    node.gpu_mem_used += gpu_mem
                node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                node.state = "Active"
                successful_jobs += 1
                results["node_states_over_time"][time][node.name] = node.state
                results["energy_consumption_over_time"][time] += node.get_power_consumption()
                break
            node_idx += 1
        else:
            original_node_obj = next((n for n in nodes if n.name == original_node), None)
            if job_type in ["AR", "IM"] and original_node_obj and original_node_obj.vm_queue and any(j[0] == "BE" for j in original_node_obj.vm_queue):
                original_node_obj.vm_queue = [j for j in original_node_obj.vm_queue if j[0] != "BE"]
                original_node_obj.cpu_used += load
                if NODE_SPEC[original_node_obj.name]["gpu_count"] > 0:
                    original_node_obj.gpu_power_used += gpu_power
                    original_node_obj.gpu_mem_used += gpu_mem
                original_node_obj.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                original_node_obj.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                successful_jobs += 1
                results["node_states_over_time"][time][original_node_obj.name] = original_node_obj.state
                results["energy_consumption_over_time"][time] += original_node_obj.get_power_consumption()

    total_time_minutes = 60
    for node in nodes:
        state_time = total_time_minutes
        results["energy_consumption"] += node.get_power_consumption() * state_time

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])

    return results

def sjf_scheduling(nodes: List[Node], workload: pd.DataFrame) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {}
    }
    total_jobs = 0
    successful_jobs = 0

    jobs = []
    times = workload["hour"].unique()
    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum()
        results["energy_consumption_over_time"][time] = 0

    for idx, row in workload.iterrows():
        time = pd.to_datetime(row["hour"])
        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        classified_jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        for job_type, load in classified_jobs:
            jobs.append((job_type, load, gpu_power, gpu_mem, time, nodename))

    total_jobs = len(jobs)
    jobs.sort(key=lambda x: x[1])

    for job in jobs:
        job_type, load, gpu_power, gpu_mem, time, original_node = job
        for node in sorted(nodes, key=lambda x: x.available_time):
            if node.can_handle_job(load, gpu_power, gpu_mem):
                node.cpu_used += load
                if NODE_SPEC[node.name]["gpu_count"] > 0:
                    node.gpu_power_used += gpu_power
                    node.gpu_mem_used += gpu_mem
                node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                node.state = "Active"
                successful_jobs += 1
                results["node_states_over_time"][time][node.name] = node.state
                results["energy_consumption_over_time"][time] += node.get_power_consumption()
                break
        else:
            original_node_obj = next((n for n in nodes if n.name == original_node), None)
            if job_type in ["AR", "IM"] and original_node_obj and original_node_obj.vm_queue and any(j[0] == "BE" for j in original_node_obj.vm_queue):
                original_node_obj.vm_queue = [j for j in original_node_obj.vm_queue if j[0] != "BE"]
                original_node_obj.cpu_used += load
                if NODE_SPEC[original_node_obj.name]["gpu_count"] > 0:
                    original_node_obj.gpu_power_used += gpu_power
                    original_node_obj.gpu_mem_used += gpu_mem
                original_node_obj.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                original_node_obj.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                successful_jobs += 1
                results["node_states_over_time"][time][original_node_obj.name] = original_node_obj.state
                results["energy_consumption_over_time"][time] += original_node_obj.get_power_consumption()

    total_time_minutes = 60
    for node in nodes:
        state_time = total_time_minutes
        results["energy_consumption"] += node.get_power_consumption() * state_time

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])

    return results

def fcfs_scheduling(nodes: List[Node], workload: pd.DataFrame) -> Dict:
    results = {
        "energy_consumption": 0,
        "job_success_rate": 0,
        "active_hosts": 0,
        "node_states_over_time": {},
        "cpu_load_over_time": {},
        "energy_consumption_over_time": {}
    }
    total_jobs = 0
    successful_jobs = 0

    jobs = []
    times = workload["hour"].unique()
    for time in times:
        results["node_states_over_time"][time] = {node.name: node.state for node in nodes}
        results["cpu_load_over_time"][time] = workload[workload["hour"] == time]["cpu_load"].sum() # 每個演算法數字都一樣
        results["energy_consumption_over_time"][time] = 0

    for idx, row in workload.iterrows():
        time = pd.to_datetime(row["hour"])
        cpu_load = row["cpu_load"]
        gpu_power = row["gpu_power_usage"]
        gpu_mem = row["gpu_mem_used"]
        nodename = row["nodename"]
        classified_jobs = classify_jobs(cpu_load, JOB_DISTRIBUTION)
        for job_type, load in classified_jobs:
            jobs.append((job_type, load, gpu_power, gpu_mem, time, nodename))

    total_jobs = len(jobs)

    for job in jobs:
        job_type, load, gpu_power, gpu_mem, time, original_node = job
        original_node_obj = next((n for n in nodes if n.name == original_node), None)
        if original_node_obj and original_node_obj.can_handle_job(load, gpu_power, gpu_mem):
            original_node_obj.cpu_used += load
            if NODE_SPEC[original_node_obj.name]["gpu_count"] > 0:
                original_node_obj.gpu_power_used += gpu_power
                original_node_obj.gpu_mem_used += gpu_mem
            original_node_obj.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
            original_node_obj.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
            original_node_obj.state = "Active"
            successful_jobs += 1
            results["node_states_over_time"][time][original_node_obj.name] = original_node_obj.state
            results["energy_consumption_over_time"][time] += original_node_obj.get_power_consumption()
        else:
            for node in sorted(nodes, key=lambda x: x.available_time):
                if node.can_handle_job(load, gpu_power, gpu_mem):
                    node.cpu_used += load
                    if NODE_SPEC[node.name]["gpu_count"] > 0:
                        node.gpu_power_used += gpu_power
                        node.gpu_mem_used += gpu_mem
                    node.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                    node.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                    node.state = "Active"
                    successful_jobs += 1
                    results["node_states_over_time"][time][node.name] = node.state
                    results["energy_consumption_over_time"][time] += node.get_power_consumption()
                    break
            else:
                if job_type in ["AR", "IM"] and original_node_obj and original_node_obj.vm_queue and any(j[0] == "BE" for j in original_node_obj.vm_queue):
                    original_node_obj.vm_queue = [j for j in original_node_obj.vm_queue if j[0] != "BE"]
                    original_node_obj.cpu_used += load
                    if NODE_SPEC[original_node_obj.name]["gpu_count"] > 0:
                        original_node_obj.gpu_power_used += gpu_power
                        original_node_obj.gpu_mem_used += gpu_mem
                    original_node_obj.update_state((time - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60, load)
                    original_node_obj.available_time = (time + timedelta(hours=1) - pd.Timestamp("2024-11-04 15:00:00")).total_seconds() / 60
                    successful_jobs += 1
                    results["node_states_over_time"][time][original_node_obj.name] = original_node_obj.state
                    results["energy_consumption_over_time"][time] += original_node_obj.get_power_consumption()

    total_time_minutes = 60
    for node in nodes:
        state_time = total_time_minutes
        results["energy_consumption"] += node.get_power_consumption() * state_time

    results["job_success_rate"] = (successful_jobs / total_jobs) * 100 if total_jobs > 0 else 0
    results["active_hosts"] = len([n for n in nodes if n.state == "Active"])

    return results

# Visualization function
def visualize_results(all_results):
    times = list(all_results["EAVMAT"]["node_states_over_time"].keys())
    times = pd.to_datetime(times)

    # 1. Node State Changes Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        state_nums = []
        for time in times:
            states = list(result["node_states_over_time"][time].values())
            active_count = sum(1 for s in states if s == "Active")
            idle_count = sum(1 for s in states if s == "Idle")
            standby_count = sum(1 for s in states if s == "Standby")
            total_nodes = len(states)
            if total_nodes > 0:
                if active_count > 0:
                    state_avg = 1
                elif standby_count > 0:
                    state_avg = 2
                else:
                    state_avg = 0
            else:
                state_avg = 0
            state_nums.append(state_avg)
        plt.plot(times, state_nums, label=f"{algo_name} Node State", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Node State (0=Idle, 1=Active, 2=Standby)')
    plt.title('Node State Changes Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 2. CPU Load Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        cpu_loads = [result["cpu_load_over_time"][t] for t in times]
        plt.plot(times, cpu_loads, label=f"{algo_name} CPU Load", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Total CPU Load')
    plt.title('CPU Load Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 3. Energy Consumption Over Time
    plt.figure(figsize=(12, 6))
    for algo_name, result in all_results.items():
        energy_values = [result["energy_consumption_over_time"][t] for t in times]
        plt.plot(times, energy_values, label=f"{algo_name} Energy", marker='o')
    plt.xlabel('Time')
    plt.ylabel('Energy Consumption (Watt-minutes)')
    plt.title('Energy Consumption Over Time by Algorithm')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    # Initialize all nodes
    nodes = [Node(name) for name in NODE_SPEC.keys()]


    workload = CPU_request_data[0:100] # pd.DataFrame(workload_data)
    workload["hour"] = pd.to_datetime(workload["hour"])

    # Run and compare all scheduling algorithms
    algorithms = {
        "EAVMAT": eavmat_scheduling,
        "PSO": pso_scheduling,
        "Round Robin": round_robin_scheduling,
        "SJF": sjf_scheduling,
        "FCFS": fcfs_scheduling
    }

    all_results = {}
    for name, algo in algorithms.items():
        for node in nodes:
            node.reset_usage()
        results = algo(nodes.copy(), workload)
        all_results[name] = results
        print(f"\n{name} Scheduling Results:")
        print(f"Total Energy Consumption (Watt-minutes): {results['energy_consumption']}")
        print(f"Job Success Rate (%): {results['job_success_rate']:.2f}")
        print(f"Number of Active Hosts: {results['active_hosts']}")
        for node in nodes:
            if node.state == "Active":
                print(f"Node {node.name} - State: {node.state}, CPU Used: {node.cpu_used}, GPU Power Used: {node.gpu_power_used}, Power: {node.get_power_consumption()} Watts")

    # Visualize results
    visualize_results(all_results)
```
