```python
import pandas as pd

def process_cpu_gpu_changes(input_file: str, output_file: str):
    # 讀取 CSV 檔案
    df = pd.read_csv(input_file)
    df['hour'] = pd.to_datetime(df['hour'])
    
    
    # 按照節點 (node) 和時間排序，確保時間順序正確
    df = df.sort_values(by=['node', 'hour'])
    
    # 計算相鄰時間的變化量
    change_columns = ['cpu_load', 'cpu_alloc', 'cpu_idle', 'cpu_total', 'gpu_power_usage', 'gpu_mem_used', 'gpu_gpu_usage', 'gpu_fan']
    for col in change_columns:
        df[f'{col}_change'] = df.groupby('node')[col].diff()
    
    # 移除第一筆變化量為 NaN 的行（因為它沒有前一筆數據可計算變化量）
    df = df.dropna()
    
    # 儲存結果到輸出 CSV 檔案
    df.to_csv(output_file, index=False)
    
    print(f"處理完成，結果已儲存到 {output_file}")

# 指定輸入與輸出文件
input_csv = "data/cpu_request_202501_hourly_node.csv"
output_csv = "data/subtract_cpu_gpu_request_changes.csv"
process_cpu_gpu_changes(input_csv, output_csv)
```
